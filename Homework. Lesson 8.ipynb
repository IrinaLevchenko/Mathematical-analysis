по заданному распределению. Построить график.

```python
import numpy as np
np.set_printoptions(precision=2, suppress=True)

from matplotlib import pylab as plt
%matplotlib inline

from scipy.optimize import fsolve
```

```python
# Задаем рандомное распределение точек
xi = np.random.uniform(-8, 8, 200)
yi = 1 / (1 + np.exp(-np.random.uniform(0.5, 1.5, 200) * xi - np.random.uniform(-1, 1, 200)))

plt.figure(figsize=(16, 8))
plt.plot(xi,yi, marker="o", ls="")
plt.show()
```

Решение:

```python
# Задаем рандомное распределение точек
xi = np.random.uniform(-8, 8, 200)
yi = 1 / (1 + np.exp(-np.random.uniform(0.5, 1.5, 200) * xi - np.random.uniform(-1, 1, 200)))

plt.figure(figsize=(16, 8))
plt.plot(xi,yi, marker="o", ls="")


def equastions(p):
    a, b = p
    exp = np.exp(-a * xi - b)
    # Производные посчитаны в тетради ((y - f(x)) ** 2)
    return ((2 * xi * (yi - 1 / (1 + exp)) * exp / (1 + exp) ** 2).sum(), 
            (2 * (yi - 1 / (1 + exp)) * exp / (1 + exp) ** 2).sum())
    
a, b = fsolve(equastions, (0,0))

x = np.linspace(-8, 8, 200)
plt.plot(x, 1 / (1 + np.exp(-a * x - b)))

plt.show()
```
